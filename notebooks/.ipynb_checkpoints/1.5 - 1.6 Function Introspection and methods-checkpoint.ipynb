### Unpacking works with any iterable type!

```python
a,b,c = 1,2,3
a
```

```python
a,b,c = "XYZ"
a, b, c
```

```python
a , b = 5, 10  #switching positions
a, b = b, a
a, b
```

```python
dictionary = {"key1": 1, "key2": 2, "key3": 3 }
a, b, c = dictionary
a, b, c
```

```python
a, b, c = dictionary
a, b, c
```

```python
def my_func(a,b,c):
    return a+b+c

my_func(*(1,2,3))
```

```python
def my_func(*args):
    cnt=0
    for arg in args:
        cnt+=arg
    return cnt

my_func(1,2,3,4), my_func(1,2,3,4,5,6)
```

```python
lst = [1,2,3,4,5]
a, *b = lst
a, b
```

### even if we start with a tuple, unpacking will create a list

```python
lst = (1,2,3,4,5)
a, *b = lst
a, b
```

```python
a, *b = "XYZW"
a, b
```

```python
a, *b, c, d = "pythonic"
a,b, c, d
```

### combining lists with `*`

```python
lst1 = [1,2,3]
lst2 = [4,5,6]

l = [*lst1, *lst2]
l
```

```python
d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

d = {*d1, *d2}
d
```

```python
d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

d = [*d1, *d2]
d
```

```python
d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

d = *d1, *d2
d
```

```python
l = [1,2,*[3,4]]
l
```

### combining dictionaries with `**`

```python
d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

d = {**d1, **d2}
d
```

### merge

```python
c = {"e":5, "f":6, **d1, **d2}
c
```

## `1.6`

```python
def add(a=1,b=2):
    return a+b

add.category = "math"
add.sub_function = lambda x,y : f"{x} + {y} = {add(x,y)}"

add(2,3)
```

```python
add.category
```

```python
add.sub_function(2,3)
```

```python
dir(add) # returns attributes for that object
```

```python
add.__name__
```

### `co_varnames` returns parameters and local variables

```python
my_func.__code__.co_varnames
```

```python
def my_func(a, b=2, c=3, *, kw1, kw2=2):
    i = 10
    b = min(i,b)
    return a * b + kw2
```
